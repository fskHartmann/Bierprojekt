```python
import pandas as pd
from tabfairgan import TFG

# Lade deinen bereinigten Datensatz
df = pd.read_csv("clean_data.csv")  
df
```

```python
# Liste an kategorischen Spalten (z.â€¯B. alles mit 'Q' auÃŸer Geburtsjahr)
categorical_cols = [col for col in df.columns if not col.startswith("Q14")]

# Wandle diese Spalten in object
df[categorical_cols] = df[categorical_cols].astype("object")
```

```python
# Zeilen mit NaN Werten entfernen
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
# Initialisiere TabFairGAN ohne Fairness Constraints
tfg = TFG(df, epochs=200, batch_size=256, device='cpu') 

# Trainiere das Modell
tfg.train()

# Generiere synthetische Daten
fake_df = tfg.generate_fake_df(num_rows=len(df))

# Speichere die synthetischen Daten ab
fake_df.to_csv("synthetic_data.csv", index=False)
```
